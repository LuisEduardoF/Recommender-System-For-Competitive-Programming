```python
import pandas as pd

df = pd.read_csv("../DATA/user_problem.csv", index_col=0)
df
```

```python
bin_edges = [-float('inf'), 1, 3, 6, 9, 12, float('inf')]
bin_labels = [0, 5, 4, 3, 2, 1]

for col in df.columns:
    df[col] = pd.cut(df[col], bins=bin_edges, labels=bin_labels, right=False, include_lowest=True)
df
```

```python
df.to_csv("../DATA/user_problem_rating.csv")
```
